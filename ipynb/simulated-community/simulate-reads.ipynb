This notebook allows for generation of a "simulated community" by slicing the Greengenes reference set based on specific primers and to a resulting amplicon length. The code is derived from an example in the [scikit-bio-cookbook](https://github.com/biocore/scikit-bio-cookbook).

````python
from os.path import join, expandvars

import pandas as pd
import numpy as np
from biom.table import Table
from biom.util import biom_open
from skbio.util import create_dir
from skbio.alignment import SequenceCollection, global_pairwise_align_nucleotide, make_identity_substitution_matrix 
from skbio.sequence import DNA
from random import random

# set to true if select tables should be written as Excel files (useful for publication)
write_xls_files = False

# set to the number of random simulated communities that should be generated
num_iterations = 10

project_dir = expandvars("$HOME/data/short-read-tax-assignment")
simulated_community_definitions = {
 'B1': (expandvars("$HOME/data/gg_13_8_otus/rep_set/97_otus.fasta"),
        expandvars("$HOME/data/gg_13_8_otus/taxonomy/97_otu_taxonomy.txt"),
        "gg_13_8_otus",
        "GTGCCAGCMGCCGCGGTAA", "ATTAGAWACCCBDGTAGTCC", "515F", "806R", np.nan),
 'B2': (expandvars("$HOME/data/gg_13_8_otus/rep_set/97_otus.fasta"),
        expandvars("$HOME/data/gg_13_8_otus/taxonomy/97_otu_taxonomy.txt"), 
        "gg_13_8_otus",
        "GTGCCAGCMGCCGCGGTAA", "ATTAGAWACCCBDGTAGTCC", "515F", "806R", 100),
 'F1': (expandvars("$HOME/data/unite-14.11/97_otus.fasta"),
        expandvars("$HOME/data/unite-14.11/97_tax.txt"),
        "unite-97-rep-set",
        "ACCTGCGGARGGATCA", "AACTTTYARCAAYGGAT", "BITSf", "B58S3r", np.nan),
 'F2': (expandvars("$HOME/data/unite-14.11/97_otus.fasta"),
        expandvars("$HOME/data/unite-14.11/97_tax.txt"),
        "unite-97-rep-set",
        "ACCTGCGGARGGATCA", "AACTTTYARCAAYGGAT", "BITSf", "B58S3r", 100),
 }
simulated_community_definitions = pd.DataFrame.from_dict(simulated_community_definitions, orient="index")
simulated_community_definitions.columns = ["Reference file path", "Reference tax path", "Reference id", "Fwd primer", "Rev primer", 
                                           "Fwd primer id", "Rev primer id", "Max length"]
simulated_community_definitions
````

````python
if write_xls_files: 
    simulated_community_definitions.to_excel(
                'simulated_community_definitions.xlsx', 
                na_rep='NA', float_format="%1.3f")
````

````python
for iteration in range(num_iterations):
    for index, data in simulated_community_definitions.iterrows():
        ## first we generate the simulated community. This includes the "representative sequences"
        ## and the biom table with no taxonomic information
        reference_fasta_fp = data['Reference file path']
        
        dataset_id = "%s-iter%d" % (index, iteration)
        simulated_community_dir = join(project_dir, "data", "simulated-community", dataset_id)
        create_dir(simulated_community_dir)
        simulated_community_rep_set_fp = join(simulated_community_dir, "rep_set.fna")
        simulated_community_biom_fp = join(simulated_community_dir, "table-no-tax.biom")
        fwd_primer = DNA(data['Fwd primer'], "fwd-primer")
        rev_primer = DNA(data['Rev primer'], "rev-primer")

        reference_sequences = SequenceCollection.read(reference_fasta_fp, constructor=DNA)
        substitution_matrix = make_identity_substitution_matrix(1, -2, DNA.alphabet())

        fwd_primer = list(fwd_primer.nondegenerates())[0]
        rev_primer = list(rev_primer.nondegenerates())[0]

        simulated_community_rep_set_f = open(simulated_community_rep_set_fp, 'w')
        otu_ids = []
        for sequence in reference_sequences:
            if random() <= 0.10:
                alignment = global_pairwise_align_nucleotide(fwd_primer, sequence, substitution_matrix=substitution_matrix)
                alignment = global_pairwise_align_nucleotide(rev_primer, alignment, substitution_matrix=substitution_matrix)
                fwd_primer_start = alignment["fwd-primer"].gap_vector().index(False)
                rev_primer_end = alignment.sequence_length() - alignment["rev-primer"].gap_vector()[-1::-1].index(False) - 1
                trimmed_sequence = sequence[fwd_primer_start:rev_primer_end]
                if len(trimmed_sequence) == 0: 
                    continue
                else:
                    otu_ids.append(trimmed_sequence.id)
                if not np.isnan(data['Max length']):
                    trimmed_sequence = trimmed_sequence[:int(data['Max length'])]
                simulated_community_rep_set_f.write(trimmed_sequence.to_fasta())

        data = np.ones((len(otu_ids), 1))
        sample_ids = ['S1']
        table = Table(data, otu_ids, sample_ids)
        with biom_open(simulated_community_biom_fp, 'w') as f:
            table.to_hdf5(f, "example")
        simulated_community_rep_set_f.close()
        
        ## next, we generate the expected results for this simulated community
        expected_dir = join(project_dir, "data", "precomputed-results/", "simulated-community", 
                          dataset_id, data["Reference id"], "expected")
        expected_biom_fp = join(output_dir, "table.biom")
        create_dir(expected_dir)
        if exists(expected_biom_fp):
            !rm $expected_biom_fp

        add_metadata_cmd = "biom add-metadata -i %s -o %s --observation-metadata-fp %s --observation-header OTUID,taxonomy --sc-separated taxonomy" \
            % (simulated_community_biom_fp, expected_biom_fp, data["Reference tax path"])
        print add_metadata_cmd
        !$add_metadata_cmd

        summarize_taxa_cmd = "summarize_taxa.py -i %s -o %s -L 2,3,4,5,6,7 --suppress_classic_table_output" % (expected_biom_fp, expected_dir)
        print summarize_taxa_cmd
        !$summarize_taxa_cmd
        for i in range(2,8):
            start_path = join(expected_dir, "table_L%d.biom" % i)
            end_path = join(expected_dir, "table.L%d-taxa.biom" % i)
            !mv $start_path $end_path
        
        ## finally, we create the reference sequences for these simulated communities, which is a version of the 
        ## reference sequences that was used to simulate the community, excluding the sequences that are in
        ## simulated_community_rep_set_fp.
        simulated_community_reference_fp = join(simulated_community_dir, "ref.fna")
        filter_fasta_cmd = "filter_fasta.py -f %s -a %s --negate -o %s" % (reference_fasta_fp, simulated_community_rep_set_fp, simulated_community_reference_fp)
        gzip_cmd = "gzip %s" % simulated_community_reference_fp
        
        print "Completed %s." % simulated_community_dir
````

```python
from glob import glob
from os.path import join, split, splitext
import pandas as pd
import seaborn as sns

input_dir = "/Users/caporaso/temp/2014.12.09-tax-assignment-runtime"
```

```python
data = []
for fp in glob(join(input_dir, "timing-results", "*.txt")):
    run_id_fields = splitext(split(fp)[1])[0].split('-')
    runtime_s = open(fp, 'U').read().strip()
    
    # DELETE THE FOLLOWING LINE WHEN ALL DATA IS READY:
    if runtime_s == "": continue
    
    data.append(('-'.join(run_id_fields[0:-4]), float(run_id_fields[-4][1:]), float(run_id_fields[-3][1:]), 
                 int(run_id_fields[-2][4:]), float(runtime_s)))
```

```python
df = pd.DataFrame(data, columns=["Method", "Query fraction", "Reference fraction", "Iteration", "Runtime"])
df = df.groupby(("Method", "Query fraction", "Reference fraction")).median().reset_index()
```

```python
data = sns.lmplot("Reference fraction", "Runtime", col="Method", 
                  data=df[df["Query fraction"] == 1.0],
                  ci=None, size=5, 
                  scatter_kws={"s": 50, "alpha": 1}, sharey=False)
```

```python
from glob import glob
from os.path import join, split, splitext
import pandas as pd
import seaborn as sns

input_dir = "/Users/caporaso/temp/2014.12.12-tax-assignment-runtime"
num_input_seqs = 9910
num_reference_seqs = 89412
```

```python
data = []
for fp in glob(join(input_dir, "timing-results", "*.txt")):
    run_id_fields = splitext(split(fp)[1])[0].split('-')
    runtime_s = open(fp, 'U').read().strip()
    
    # DELETE THE FOLLOWING LINE WHEN ALL DATA IS READY:
    if runtime_s == "": continue
    
    data.append((run_id_fields[0], 
                float(run_id_fields[1][1:]) * num_input_seqs, 
                float(run_id_fields[2][1:]) * num_reference_seqs, 
                int(run_id_fields[3][4:]),
                float(runtime_s)))
```

```python
df = pd.DataFrame(data, columns=["Method", "Number of Query Sequences", "Number of Reference Sequences", "Iteration", "Runtime"])
df = df.groupby(("Method", "Number of Query Sequences", "Number of Reference Sequences")).median().reset_index()

# DROP THIS LINE WHEN BLAST COMPLETES
#df = df[df["Method"] != 'blast']
```

```python
pl = sns.lmplot("Number of Reference Sequences", "Runtime", col="Method", 
                  data=df[df["Number of Query Sequences"] == num_input_seqs],
                  ci=None, size=5, 
                  scatter_kws={"s": 50, "alpha": 1}, sharey=True)
```

```python
pl = sns.lmplot("Number of Query Sequences", "Runtime", col="Method", 
                data=df[df["Number of Reference Sequences"] == num_reference_seqs],
                ci=None, size=5, 
                scatter_kws={"s": 50, "alpha": 1}, sharey=True)
```

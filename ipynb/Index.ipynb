# An extensible framework for optimizing classification enhances marker-gene taxonomic assignments: Supplementary Notebooks

These notebooks were used to perform the analyses presented in (Bokulich, Rideout, et al. (in preparation)), and can be used to reproduce the analyses in that paper, or to extend them to other data sets. 

To run any of the analysis notebooks, you'll need the [tax-credit project](https://github.com/gregcaporaso/short-read-tax-assignment/). See the [README](https://github.com/gregcaporaso/short-read-tax-assignment/blob/master/README.md) for installation instructions. For a static version of these notebooks that you can view as a webpage without installing anything, you can [view the notebooks on nbviewer](http://nbviewer.ipython.org/github/gregcaporaso/short-read-tax-assignment/tree/master/ipynb/Index.ipynb).

Questions should be posted as issues on the [tax-credit repository issue tracker](https://github.com/gregcaporaso/short-read-tax-assignment/issues)

## Extending the analyses to include evaluation of new taxonomic assignment methods

Given a set of precomputed taxonomic assignment method evaluation results (this will likely be the ones included in the [tax-credit GitHub repository](https://github.com/gregcaporaso/short-read-tax-assignment/)) and optionally a set of results generated by the user, the results generated by the user can be analyzed in the context of the precomputed results. This allows users to rapidly determine how a new method performs, relative to the precomputed results. Additionally, users can submit their taxonomic assignment method results to the [tax-credit repository](https://github.com/gregcaporaso/short-read-tax-assignment/) as a pull request so they can be included as precomputed results for future users.
